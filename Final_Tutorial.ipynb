### Tidying "Sleep Health and Lifestyle Dataset" Dataset
We will clean up the dataset by ridding it of any unnecessary columns, renaming and rearranging columns, and converting specific value types as one-hot encodings for improved analysis on the data.

Fortunately, no missing data was detected from the dataset, therefore we will not have to impute missing values into the data.

```python
# Make one-hot encoding for the different occupations in the dataset
sleep_health_and_lifestyle_df_tidy = pd.get_dummies(sleep_health_and_lifestyle_df, columns=['Occupation'], drop_first=True)

# Encode Gender columns
sleep_health_and_lifestyle_df_tidy['Gender'] = sleep_health_and_lifestyle_df['Gender'].map({'Male': 1, 'Female': 0})

# Convert numerical data types into floats for easier use later on
features_to_convert = ['Age', 'Physical Activity Level', 'Stress Level', 'Daily Steps', 'Sleep Duration']
sleep_health_and_lifestyle_df_tidy[features_to_convert] = sleep_health_and_lifestyle_df_tidy[features_to_convert].astype(float)

# Drop unnecessary columns from the dataset and add original 'Occupation' column from original dataset
features_to_drop = ['Quality of Sleep', 'BMI Category', 'Blood Pressure', 'Heart Rate', 'Sleep Disorder', 'Daily Steps']
sleep_health_and_lifestyle_df_tidy.drop(columns=features_to_drop, inplace=True)
sleep_health_and_lifestyle_df_tidy['Occupation'] = sleep_health_and_lifestyle_df['Occupation']

# Rename appropriate columns for clarity
feature_renaming = {
    'Person ID': 'Person_ID',
    'Sleep Duration': 'Sleep_Duration',
    'Physical Activity Level': 'Physical_Activity_Level',
    'Stress Level': 'Stress_Level',
    'Occupation_Software Engineer': 'Occupation_Software_Engineer',
    'Occupation_Sales Representative': 'Occupation_Sales_Representative'
}
sleep_health_and_lifestyle_df_tidy.rename(columns=feature_renaming, inplace=True)

# Rearrange columns for clarity
sleep_health_and_lifestyle_df_tidy = sleep_health_and_lifestyle_df_tidy[['Person_ID', 'Gender', 'Age', 'Sleep_Duration', 'Physical_Activity_Level', 'Occupation', 'Occupation_Doctor', 'Occupation_Engineer', 'Occupation_Lawyer', 'Occupation_Manager', 'Occupation_Nurse', 'Occupation_Sales_Representative', 'Occupation_Salesperson', 'Occupation_Scientist', 'Occupation_Software_Engineer', 'Occupation_Teacher', 'Stress_Level']]

sleep_health_and_lifestyle_df_tidy.head(10)
```

<h2><center>Hypothesis Testing and Data Modeling for "Sleep Health and Lifestyle Dataset" Dataset</center></h2>

We will proceed to make a regression models that forecast stress levels based on our extensive dataset. But before we put it to use, it's a good idea to examine the accuracy of the data using a Z-test. This kind of examination is known as hypothesis testing.

**Null Hypothesis**

The null hypothesis in this case claims that factors like gender, age, the level of physical activity, and sleep duration don't actually impact stress levels.

```python
# Statsmodels ols to get statistic regarding different factors and stress level. 
formula = "Stress_Level ~ Gender + Age + Sleep_Duration + Physical_Activity_Level"
sleep_health_lifestyle_stats = stm.ols(formula=formula, data=sleep_health_and_lifestyle_df_tidy).fit()
sleep_health_lifestyle_stats.summary()
```

The calculated Prob (F-statistic) value of 1.00e-114 indicates that there is a significant relationship between the combined set of features and sleep duration. This extremely low p-value suggests that at least one of these features has a non-zero coefficient in the linear regression model, indicating a strong overall statistical significance.


**Rejecting the Null Hypothesis**

Among the individual features, gender, sleep duration, and the level of physical activity exhibit highly significant impacts on the stress levels for the entire dataset, as indicated by their very low P>|t| values of 0.000. These results suggest that higher levels of physical activity, increased sleep durations, and gender are associated with changes in sleep duration.

However, among them Gender does not seem to make much sense. With a standard error of 0.113, it is the highest of any of the features. Thus, it appears to be more of a feature by chance.

```python
# Parameters from the stats model above. This can help us understand which factor contributes most to stress.
sleep_health_lifestyle_stats.params
```

From these coefficients data, however, we can observe the following:
1. Only sleep duration is negatively correlated with stress levels. Decreases in sleep duration will lead to a possible increase in stress levels.
2. Stress level has the strongest correlation with sleep duration, indicating that it truly has more of an impact on it (although negative).
3. Gender has a strong postive correlation with stress levels, however it has a fairly high standard error. This means that there is bound to be inaccuracies in the data.

Thus, we are rejecting the null hypothesis that none of the considerable features have an impact on stress levels.

Therefore, we can construct Machine Learning models for this purpose. Once we've trained the model, we can visualize the coefficients of the four key features to understand their influence on the final prediction.

We will compare the linear and logisitic regression models to receive an accuracy.

**Linear Regression and Logisistic Regression for Stress Level Forecast**

```python
# Select features
selected_features = ["Gender", "Age", "Sleep_Duration", "Physical_Activity_Level"]
feature_indices = [sleep_health_and_lifestyle_df_tidy.columns.get_loc(col) for col in selected_features]

# Split the data into the training and test data
X = sleep_health_and_lifestyle_df_tidy.values[:, feature_indices]
y = sleep_health_and_lifestyle_df_tidy.values[:, -1].astype(float)

# Before fitting the data, we need to implement feature scaling as the four types have different ranges and distributions of data
# Thus, we will standardize the features (mean = 0, std = 1)
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.3, random_state=50)

# Compute the Linear Regression Model
linear_regression_model = linear_model.LinearRegression()
linear_regression_model = linear_regression_model.fit(X_train, y_train)

# Calculate MSE
linear_regression_predictions = linear_regression_model.predict(X_test)
linear_regression_mse = mean_squared_error(y_test, linear_regression_predictions)
print("Linear Regression MSE:", linear_regression_mse)

# Compute the Logisitic Regression Model
logistic_regression_model = linear_model.LogisticRegression()
logistic_regression_model.fit(X_train, y_train)

# Calculate the accuracy of the model
logistic_regression_predictions = logistic_regression_model.predict(X_test)
print("Logistic Regression Accuracy: ", logistic_regression_model.score(X_test, y_test))

# Plot predicted vs. actual values from the linear regression
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.regplot(x=y_test, y=linear_regression_predictions, ax=axes[0])
axes[0].set_title('Linear Regression: Predicted vs. Test')
axes[0].set_xlabel('Predicted Values')
axes[0].set_ylabel('Test Values')

sns.regplot(x=y_test, y=logistic_regression_predictions, ax=axes[1])
axes[1].set_title('Logistic Regression: Predicted vs. Test')
axes[1].set_xlabel('Predicted Values')
axes[1].set_ylabel('Test Values')

plt.tight_layout()
plt.show()
```

The Linear Regression MSE value is posed to be 0.86, which indicates that, on average, the squared difference between the predicted stress levels and the actual stress levels is relatively small (compared to the scaling of the data). This suggests that the Linear Regression model is doing a decent job in approximating the stress levels. 
On the other hand, the accuracy of the logistic regression model is fairly high (.85). This suggests that the Logisitic Regression model is doing well in approximating stress levels.
The positive slope in both the scatter plots suggest that the models' predictions increase as the stress level increases.

To better understand the data, we can determine how valuable each feature is. Feature importance indicates the usefulness of each feature in improving the model's accuracy. There are a few ways to indicate feature importance, such as Random Forest Classification. However, Random Forest Classification seems overkill for this fairly simple linear data. Thus, we will stick to the coefficients in the Linear Regression from the predicted data to determine feature importance.

```python
# Find the feature importance vector by taking the coefficients from the Linear Regression
feature_importance = pd.Series(linear_regression_model.coef_, index=selected_features)

# Sort features by importance
sorted_importance = feature_importance.abs().sort_values(ascending=False)

print("Feature Importance:")
print(sorted_importance)

plt.figure(figsize=(10, 6))
sns.barplot(x=sorted_importance.values, y=sorted_importance.index, palette='viridis')
sns.barplot(x=sorted_importance.values, y=sorted_importance.index, palette='viridis')
plt.xlabel('Feature Importance (Absolute Coefficient)')
plt.ylabel('Features')
plt.title('Linear Regression: Feature Importance')
plt.show()
```

We can view that sleep duration is of most importance in accordance to the predicted data. The magnitude of the coefficient (the absolute value) indicates the strength of the relationship. A larger coefficient magnitude implies a stronger impact of the feature on the target variable. In this case, a coefficient value of 1.4041 suggests that sleep duration has a relatively significant influence on predicting stress level among any occupation. It is worth noting that this is based off the whole dataset.

It is more accurate to find the feature importance in regards to each type of occupation.

```python
# In accordance to the dataset, there are 11 unqiue occupations
# Create a 2D importance array for the each occupation for every feature
feature_importance_matrix = []

# Iterates through every country once
for job in sleep_health_and_lifestyle_df_tidy['Occupation'].unique():
    
    # Get the features and the life expectancy data for the country for all 17 years
    X = sleep_health_and_lifestyle_df_tidy[sleep_health_and_lifestyle_df_tidy['Occupation'] == job].values[:, feature_indices]
    y = sleep_health_and_lifestyle_df_tidy[sleep_health_and_lifestyle_df_tidy['Occupation'] == job].values[:, -1].astype(float)

    # Compute the Linear Regression Model.
    linear_regression_model = linear_model.LinearRegression()
    linear_regression_model = linear_regression_model.fit(X, y)
    
    feature_importance = pd.Series(linear_regression_model.coef_, index=selected_features)
    feature_importance_matrix.append(linear_regression_model.coef_)
feature_importance_matrix = np.array(feature_importance_matrix)

mean_feature_importance = []
mean_feature_importance.append(abs(feature_importance_matrix[:, 0].mean()))
mean_feature_importance.append(abs(feature_importance_matrix[:, 1].mean()))
mean_feature_importance.append(abs(feature_importance_matrix[:, 2].mean()))
mean_feature_importance.append(abs(feature_importance_matrix[:, 3].mean()))

plt.figure(figsize=(10, 6))
sns.barplot(x=mean_feature_importance, y=[col for col in sleep_health_and_lifestyle_df_tidy.columns if col in ['Gender', 'Age', 'Sleep_Duration', 'Physical_Activity_Level']], palette='viridis')
plt.xlabel('Mean Feature Importance (Absolute)')
plt.ylabel('Features')
plt.title('Linear Regression: Feature Importance Mean between Occupations')
plt.show()
mean_feature_importance
```

From the mean feature importance between each occupation, we can still see that sleep duration is of highest importance with a mean feature importance magnitude of .076, for determing stress level.
